# Expectation-Maximization Functions & Class

```python
def gaussian_likelihood(x: np.ndarray, means: np.ndarray, sigmas: np.ndarray) -> np.ndarray:
    n_components, _ = means.shape
    likelihood = [multivariate_normal.pdf(
        x, means[i, :], sigmas[i, :, :], allow_singular=True) for i in range(n_components)]
    return np.asarray(likelihood).T


def min_max_norm(
    img: np.ndarray, max_val: int = None, mask: np.ndarray = None, dtype: str = None
):
    if mask is not None:
        img_ = img.copy()
        img = img[mask != 0].flatten()
    if max_val is None:
        max_val = np.iinfo(img.dtype).max
    img = (img - img.min()) / (img.max() - img.min()) * max_val
    if mask is not None:
        img_[mask != 0] = img.copy()
        img = img_.copy()
    if dtype is not None:
        return img.astype(dtype)
    else:
        return img
    
    
class ExpectationMaximization():
    def __init__(
        self,
        n_components: int = 3,
        mean_init: Union[str, np.ndarray] = 'random',
        priors: Union[str, np.ndarray] = 'non_informative',
        max_iter: int = 50,
        change_tol: float = 1e-6,
        seed: float = 420,
        verbose: bool = True,
        plot_rate: int = None,
        tissue_models: np.ndarray = None,
        atlas_use: str = None,
        atlas_map: np.ndarray = None,
        previous_result: np.ndarray = None
    ):
        self.n_components = n_components
        self.mean_init = mean_init
        self.priors = priors
        self.change_tol = change_tol
        self.previous_result = previous_result
        self.max_iter = max_iter
        self.seed = seed
        self.verbose = verbose
        self.plot_rate = plot_rate
        self.fitted = False
        self.cluster_centers_ = None
        self.n_iter_ = 0
        self.tissue_models = tissue_models
        self.atlas_use = atlas_use
        self.atlas_map = atlas_map
        if self.atlas_map is not None:
            self.atlas_map = self.atlas_map.T
            self.atlas_map[np.sum(self.atlas_map, axis=1) == 0, :] = np.ones((1, 3))/3
            epsilon = 1e-8 
            self.atlas_map = self.atlas_map / (self.atlas_map.sum(axis=1)[:, None] + epsilon)
        self.training = False

        mean_options = ['random', 'kmeans', 'tissue_models', 'label_prop']
        condition_one = isinstance(mean_init, str) and (mean_init not in mean_options)
        condition_two = isinstance(priors, np.ndarray) and (priors.size != n_components)
        

    def fit(self, x: np.ndarray):
        self.training = True
        self.x = x
        self.n_feat = x.shape[1] if np.ndim(x) > 1 else 1
        self.n_samples = len(x)
        self.labels = np.ones((self.n_samples, self.n_components))
        self.posteriors = np.zeros((self.n_samples, self.n_components))

        # Define kind of means to be used
        self.mean_type = 'Passed array'
        if isinstance(self.mean_init, str):
            if self.mean_init == 'random':
                self.mean_type = 'Random Init'
                rng = np.random.default_rng(seed=self.seed)
                idx = rng.choice(self.n_samples, size=self.n_components, replace=False)
                self.posteriors[idx, np.arange(self.n_components)] = 1
                self.priors = np.ones((self.n_components, 1)) / self.n_components
            elif self.mean_init == 'kmeans':
                self.mean_type = 'K-Means'
                kmeans = KMeans(
                    n_clusters=self.n_components, random_state=self.seed, n_init=10).fit(self.x)
                self.posteriors[np.arange(self.n_samples), kmeans.labels_] = 1
            elif self.mean_init == 'tissue_models':
                self.mean_type = 'Tissue Models'
                tissue_prob_maps = np.zeros((self.n_samples, self.n_components))
                for c in range(self.n_components):
                    tissue_prob_maps[:, c] = self.tissue_models[c, :][self.x[:, 0]]
                self.posteriors = tissue_prob_maps
                self.posteriors[np.arange(self.n_samples), np.argmax(tissue_prob_maps, axis=1)] = 1
            else:
                self.mean_type = 'Label Propagation'
                self.posteriors = self.atlas_map
                
        else:
            self.means = self.mean_init
            idx = rng.choice(self.n_samples, size=self.n_components, replace=False)
            self.labels[idx, np.arange(self.n_components)] = 1

        if self.previous_result is not None:
            self.posteriors = self.previous_result
            if self.mean_init == 'kmeans':
                self.match_labels()
            self.posteriors = self.posteriors * self.atlas_map
            self.maximization(initial=True, initial_random=(self.mean_init == 'random'))
        else:
            self.maximization(initial=True, initial_random=(self.mean_init == 'random'))

            # Expectation Maximization process
            self.expectation_maximization()
        self.cluster_centers_ = self.means
        self.fitted = True

    def predict(self, x: np.ndarray) -> np.ndarray:
        self.training = False
        self.x = x
        self.expectation()
        self.predictions = np.argmax(self.posteriors, 1)
        return self.posteriors, self.predictions

    def predict_proba(self, x: np.ndarray) -> np.ndarray:
        self.x = x
        self.training = False
        self.expectation()
        return self.posteriors

    def fit_predict(self, x: np.ndarray) -> np.ndarray:
        self.fit(x)
        self.posteriors, self.predictions = self.predict(self.x)
        return self.posteriors, self.predictions

    def expectation_maximization(self):
        prev_log_lkh = 0
        for it in tqdm(range(self.max_iter), disable=self.verbose, desc=f'EM-{self.mean_init}-{self.atlas_use}'):
            self.n_iter_ = it + 1

            # E-step
            self.expectation(it)

            # Check stop criteria: tolerance over log-likelihood
            for i in range(self.n_components):
                self.likelihood[:, i] = self.likelihood[:, i] * self.priors[i]
            log_lkh = np.sum(np.log(np.sum(self.likelihood, 1)), 0)
            difference = abs(prev_log_lkh - log_lkh)
            prev_log_lkh = log_lkh
            
            # M-Step
            self.maximization()
            if self.verbose:
                logging.info(f'Iteration: {it} - Log likelihood change: {difference}')

    def expectation(self, it: int = None):
        self.likelihood = gaussian_likelihood(self.x, self.means, self.sigmas)
        num = np.asarray([
            self.likelihood[:, j] * self.priors[j] for j in range(self.n_components)]).T
        denom = np.sum(num, 1)
        self.posteriors = np.asarray([num[:, j] / denom for j in range(self.n_components)]).T

        if (self.atlas_use == 'into' and self.training):
            self.posteriors = self.posteriors * self.atlas_map


    def match_labels(self):
        labels_em = np.argmax(self.posteriors, axis=1)
        labels_atlas = np.argmax(self.atlas_map, axis=1)
        order = {}
        for label in [1, 2]:
            labels, counts = np.unique(labels_em[labels_atlas == label], return_counts=True)
            order[label] = labels[np.argmax(counts)]

        order[0] = [i for i in [0, 1, 2] if i not in list(order.values())][0]

        # resort
        self.posteriors_ = self.posteriors.copy()
        for key, val in order.items():
            self.posteriors_[key, :] = self.posteriors[val, :]

    def maximization(self, initial: bool = False, initial_random: bool = False):

        self.labels = np.zeros((self.x.shape[0], self.n_components))
        self.labels[np.arange(self.n_samples), np.argmax(self.posteriors, axis=1)] = 1

        # Get means
        self.counts = np.sum(self.posteriors, 0)
        weithed_avg = np.dot(self.posteriors.T, self.x)

        # Get means
        self.means = weithed_avg / self.counts[:, np.newaxis]

        # Get covariances
        self.sigmas = np.zeros((self.n_components, self.n_feat, self.n_feat))
        epsilon = 1e-6  # Small value to add to the diagonal of covariance matrix   ********
        if initial_random:
            sigma = np.cov((self.x - np.mean(self.x, axis=0)).T)
            for i in range(self.n_components):
                self.sigmas[i] = sigma
                self.sigmas[i].flat[:: self.n_feat + 1] += 1e-6
            if np.ndim(self.sigmas) == 1:
                self.sigmas = (self.sigmas[:, np.newaxis])[:, np.newaxis]
        else:
            for i in range(self.n_components):
                diff = self.x - self.means[i, :]
                weighted_diff = self.posteriors[:, i][:, np.newaxis] * diff
                self.sigmas[i] = np.dot(weighted_diff.T, diff) / self.counts[i]
                # Add a small value to the diagonal of the covariance matrix
                self.sigmas[i] += np.eye(self.n_feat) * epsilon
                self.priors = self.counts / len(self.x)
                    
                       
def match_pred_w_gt(
    prediction: np.ndarray, gt_mask: np.ndarray, t1_array: np.ndarray, t2_array: np.ndarray = None
) -> np.ndarray:
    # Initialize the container
    matched_predictions = np.zeros_like(prediction)
    original_shape = prediction.shape

    # Reshape the images for cleaner code
    prediction = prediction.flatten()
    gt_mask = gt_mask.flatten()

    if t2_array is not None:
        data = np.array([t1_array.flatten(), t2_array.flatten()]).T
        n_features = 2
    else:
        n_features = 1
        data = t1_array.flatten()
        data = data.reshape(data.shape[0], -1)

    # Get the available labels
    predicted_labels = np.unique(prediction[prediction != 0])
    n_components = len(predicted_labels)

    # Get means of tissue intensities in each class under gt or prediction masks
    means_pred = np.zeros((n_components, n_features))
    means_gt = np.zeros((n_components, n_features))
    for label in range(n_components):
        means_pred[label, :] = np.median(data[prediction == (label+1), :], axis=0)
        means_gt[label, :] = np.median(data[gt_mask == (label+1), :], axis=0)

    # Compare the mean intensity of each class (euclidean distance)
    dists = np.zeros((n_components, n_components))
    for i in range(n_components):
        for j in range(n_components):
            dists[i, j] = np.linalg.norm(means_pred[i, :] - means_gt[j, :])

    # Match classes according to closer mean intensity value
    gt_match = np.argmin(dists, axis=1)
    prediction = prediction.reshape(original_shape)

    # Relabel image
    for i in range(n_components):
        matched_label = gt_match[i] + 1
        matched_predictions[prediction == i+1] = matched_label

    return matched_predictions


def match_labels(seg: np.ndarray, gt: np.ndarray, prob_array: np.ndarray = None):
    shape = seg.shape
    seg = seg.flatten()
    gt = gt.flatten()
    order = {}
    for label in [0, 2, 3]:
        labels, counts = np.unique(seg[gt == label], return_counts=True)
        order[label] = labels[np.argmax(counts)]
    order[1] = [i for i in [0, 1, 2, 3] if i not in list(order.values())][0]
    seg_ = seg.copy()
    prob_array_ = prob_array.copy() if prob_array is not None else None
    for des_val, seg_val in order.items():
        seg[seg_ == seg_val] = des_val
        if des_val in [1, 2, 3]:
            if prob_array_ is not None:
                prob_array[:, des_val-1] = prob_array_[:, seg_val-1]
    if prob_array is not None:
        return seg.reshape(shape), prob_array
    else:
        return seg.reshape(shape)


def segmentation_em(
    t1: np.ndarray, brain_mask: np.ndarray, tissue_models: np.ndarray = None,
    mean_init: str = 'kmeans', atlas_use: str = None, atlas_map: np.ndarray = None,
    previous_result: np.ndarray = None
):
    # Define data
    t1_vector = t1[brain_mask == 255].flatten()
    if atlas_map is not None:
        #print(atlas_map.shape)
        atlas_map_vector = atlas_map[:, brain_mask == 255].reshape(atlas_map.shape[0], -1)
        atlas_map_vector = atlas_map_vector[1:, :]
    else:
        atlas_map_vector = None
    data = np.array(t1_vector)[:, np.newaxis]

    model = ExpectationMaximization(
        n_components=3,
        mean_init=mean_init,
        verbose=False,
        plot_rate=None,
        tissue_models=tissue_models,
        atlas_use=atlas_use,
        atlas_map=atlas_map_vector,
        previous_result=previous_result
    )


    start = time.time()
    preds_prob, preds_categorical = model.fit_predict(data)
    t1_time = time.time() - start
    predictions_categorical = brain_mask.flatten()
    predictions_categorical[predictions_categorical == 255] = preds_categorical + 1
    t1_seg_res_cat = predictions_categorical.reshape(t1.shape)
    t1_iters = model.n_iter_
    return preds_prob, t1_seg_res_cat, t1_time, t1_iters


```

# Other Functions

```python
def dice_score(groundtruth, segmented_image):
    
    unique_classes = np.unique(groundtruth[groundtruth != 0])
    dice_scores = np.zeros(len(unique_classes))
    for class_idx, class_id in enumerate(unique_classes):
        seg_mask = np.where(segmented_image == class_id, 1, 0)
        groundtruth_mask = np.where(groundtruth == class_id, 1, 0)

        intersection = np.sum(seg_mask & groundtruth_mask)
        union = np.sum(seg_mask) + np.sum(groundtruth_mask)
        if union == 0:
            dice_scores[class_idx] = 1.0  
        else:
            dice = 2.0 * intersection / union
            dice_scores[class_idx] = round(dice, 4)
            
    return dice_scores.tolist()

def save_segementations(
    volume: np.ndarray, reference: sitk.Image, filepath: Path
):
    if (type(volume) == list) or (len(volume.shape) > 3):
        if type(volume[0]) == sitk.Image:
            vol_list = [vol for vol in volume]
        else:
            vol_list = [sitk.GetImageFromArray(vol) for vol in volume]
        joiner = sitk.JoinSeriesImageFilter()
        img = joiner.Execute(*vol_list)
    else:
        img = sitk.GetImageFromArray(volume)
    img.SetDirection(reference.GetDirection())
    img.SetOrigin(reference.GetOrigin())
    img.SetSpacing(reference.GetSpacing())
    for key in reference.GetMetaDataKeys():
        img.SetMetaData(key, reference.GetMetaData(key))
    sitk.WriteImage(img, filepath)
    
    
def segmentation_tissue_model(test_image, test_mask, tissue_model):   
    
    masked_test_image_data = test_image[test_mask == 1].flatten()
    n_classes              = tissue_model.shape[0]
    preds                  = np.zeros((n_classes, len(masked_test_image_data)))

    for c in range(n_classes):
        preds[c, :] = tissue_model[c, masked_test_image_data]

    preds = np.argmax(preds, axis=0)
    predictions = test_mask.flatten()
    predictions[predictions == 1] = preds + 1
    segmented_test_image = predictions.reshape(test_image.shape)
    
    return segmented_test_image
```

# Get Tissue Model From Training Data

```python
# Define the path of the registarted traing images
MAIN_PATH = Path('../Lab_03/').resolve()
train_outputs_dir_path      = MAIN_PATH / 'data' / 'training-set' / 'training-outputs' / 'Par0009.affine'
train_reg_images_dir_path   = train_outputs_dir_path / 'training-images'
train_reg_labels_dir_path   = train_outputs_dir_path / 'training-labels'

# Define the path of the generated atlas files usnig the training data 
atlases_dir_path            = MAIN_PATH / 'data' / 'training-set' / 'atlases' / 'Par0009.affine'

# Define the reference image
ref_image_path = train_reg_images_dir_path / 'r_1001.nii.gz'

# Import the topological atlas 
t_atlas      = sitk.GetArrayFromImage(sitk.ReadImage(str(atlases_dir_path/'t_atlas.nii.gz')))
gt_labels    = sitk.GetArrayFromImage(sitk.ReadImage(str(train_reg_labels_dir_path/'r_1001_3C.nii.gz')))

# Define the number of classes 
num_classes = 4
labels_keys = {0: 'background', 1: 'csf', 2: 'wm', 3: 'gm'}

t_histograms = np.zeros((4, 255))
gt_histograms = np.zeros((4, 255))

# Iterate over the trianing images
for i, image_path in tqdm(enumerate(train_reg_images_dir_path.iterdir()), total=len(list(train_reg_images_dir_path.iterdir()))):
    if image_path.name.endswith('txt'):
        continue
    # Get the name of the moving image
    moving_image_id   = image_path.name.rstrip('.nii.gz')
    
    # Get the path of the moving label
    moving_label_path = train_reg_labels_dir_path / f'{moving_image_id}_3C.nii.gz'
    
    # Read the image and convert into array
    image_array = sitk.GetArrayFromImage(sitk.ReadImage(str(image_path)))
    image_array = ((image_array - np.min(image_array)) / (np.max(image_array) - np.min(image_array))) * 255
    image_array = image_array.astype(np.uint8)
    
    # Read the laebl
    image_label = sitk.GetArrayFromImage(sitk.ReadImage(str(moving_label_path)))

    for j in range(num_classes):
        t_histograms[j, :]  += np.histogram(image_array[t_atlas == j], bins=255, range=[0, 256])[0]
        gt_histograms[j, :] += np.histogram(image_array[image_label == j], bins=255, range=[0, 256])[0]

t_histograms_density  = t_histograms / np.sum(t_histograms, axis=1)[:, None]
gt_histograms_density = gt_histograms / np.sum(gt_histograms, axis=1)[:, None]

t_sum  = np.sum(t_histograms_density, axis=0)[None, :]
gt_sum = np.sum(gt_histograms_density, axis=0)[None, :]

t_histograms_posterior  = t_histograms_density / np.repeat(t_sum, 4, axis=0)
gt_histograms_posterior = gt_histograms_density / np.repeat(gt_sum, 4, axis=0)
```

```python
# Define the path of the test images, labels, masks and outputs
MAIN_PATH = Path('../Lab_03/').resolve()
test_set_dir_path     = MAIN_PATH / 'data' / 'test-set'
test_images_dir_path  = test_set_dir_path / 'testing-images'
test_labels_dir_path  = test_set_dir_path / 'testing-labels'
test_masks_dir_path   = test_set_dir_path / 'testing-masks'
test_outputs_dir_path = test_set_dir_path / 'testing-outputs'


# Define the path of the tissue probability map
tissue_prob_maps_path_ours = test_outputs_dir_path / 'testing-our-atlas'


# Define the tissue model for segmentation
tissue_models = gt_histograms_posterior
tissue_models[:, np.argmax(tissue_models, 0) == 0] = np.ones((4, 1))
tissue_models = tissue_models[1:, :]
tissue_models = tissue_models / np.sum(tissue_models, 0)[None, :]


#Define sum blank list to store segmented images
segmented_images_em_kmean   = []
segmented_images_em_tm      = []
segmented_images_em_lp      = []

segmented_images_em_kmean_into = []
segmented_images_em_tm_into    = []
segmented_images_em_lp_into    = []

em_kmean_dice_scores = []
em_tm_dice_scores    = []
em_lp_dice_scores    = []

em_kmean_into_dice_scores = []
em_tm_into_dice_scores    = []
em_lp_into_dice_scores    = []

# Define lists to store computation time and iteration
times_em_kmean = []
times_em_tm    = []
times_em_lp    = []

times_em_kmean_into = []
times_em_tm_into    = []
times_em_lp_into    = []

iterations_em_kmean = []
iterations_em_tm    = []
iterations_em_lp    = []

iterations_em_kmean_into = []
iterations_em_tm_into    = []
iterations_em_lp_into    = []


# Iterate over the test images for the segmentation using tissue model
for k, test_image_path in enumerate(list(test_images_dir_path.iterdir())):
    test_image_name = test_image_path.name.rstrip('.nii.gz')

    # Load test image, label, mask and tissue probability map
    test_image = sitk.ReadImage(str(test_image_path))
    test_gt    = sitk.ReadImage(str(test_labels_dir_path / f'{test_image_name}_3C.nii.gz'))
    test_mask  = sitk.ReadImage(str(test_masks_dir_path / f'{test_image_name}_1C.nii.gz'))
    tissue_prob_map_ours = sitk.ReadImage(str(tissue_prob_maps_path_ours / f'{test_image_name}_atlas.nii.gz'))
    
    # Convert test image, label, and mask into arrays
    test_image_array = sitk.GetArrayFromImage(test_image)
    test_image_array = ((test_image_array - np.min(test_image_array)) / (np.max(test_image_array) - np.min(test_image_array)) * 255).astype(np.uint8)
    test_gt_array    = sitk.GetArrayFromImage(test_gt).astype(np.uint8)
    test_mask_array  = sitk.GetArrayFromImage(test_mask)
    
    tissue_prob_map_ours = sitk.GetArrayFromImage(tissue_prob_map_ours)
    tissue_prob_map_ours = np.where(tissue_prob_map_ours > 1, 1, tissue_prob_map_ours)
    tissue_prob_map_ours = np.where(tissue_prob_map_ours < 0, 0, tissue_prob_map_ours)

    # Read the upadated mask and convert into array 
    new_mask         = sitk.ReadImage(str(test_set_dir_path / 'new-masks'/ f'{test_image_name}brain_mask.nii'))
    new_mask_array   = sitk.GetArrayFromImage(new_mask) 
    
    
    #**---------------------SEGMENTATION USING EM & KMEANS-----------------------------**#
    prob_em_kmean, segmented_image_em_kmean_, time_em_kmean, iteration_em_kmean = segmentation_em(test_image_array, new_mask_array, None, 'kmeans')
    segmented_image_em_kmean, prob_em_kmean = match_labels(segmented_image_em_kmean_, test_gt_array, prob_em_kmean)
    
    segmented_images_em_kmean.append(segmented_image_em_kmean)
    times_em_kmean.append(time_em_kmean)
    iterations_em_kmean.append(iteration_em_kmean)
    
    
    #**---------------------SEGMENTATION USING EM & TISSUE MODEL------------------------**#
    prob_em_tm, segmented_image_em_tm, time_em_tm, iteration_em_tm = segmentation_em(test_image_array, new_mask_array, tissue_models, 'tissue_models')
    
    segmented_images_em_tm.append(segmented_image_em_tm)
    times_em_tm.append(time_em_tm)
    iterations_em_tm.append(iteration_em_tm)
    
    
    #**---------------------SEGMENTATION USING EM & LABEL PROPAGATION-------------------**#
    prob_em_lp, segmented_image_em_lp, time_em_lp, iteration_em_lp = segmentation_em(test_image_array, new_mask_array, tissue_models, 'label_prop', None, tissue_prob_map_ours)
    
    segmented_images_em_lp.append(segmented_image_em_lp)
    times_em_lp.append(time_em_lp)
    iterations_em_lp.append(iteration_em_lp)
    

    
    # -------------------------------- SEGMENTATION USING EM ATLAS INTO--------------------------------------**#
    _, segmented_image_em_kmean_into, time_em_kmean_into, iteration_em_kmean_into = \
        segmentation_em(test_image_array, new_mask_array, None, 'kmeans', 'into', tissue_prob_map_ours)
    
    segmented_images_em_kmean_into.append(segmented_image_em_kmean_into)
    times_em_kmean_into.append(time_em_kmean_into)
    iterations_em_kmean_into.append(iteration_em_kmean_into)
 
    _, segmented_image_em_tm_into, time_em_tm_into, iteration_em_tm_into = \
        segmentation_em(test_image_array, new_mask_array, tissue_models, 'tissue_models', 'into', tissue_prob_map_ours)
    
    segmented_images_em_tm_into.append(segmented_image_em_tm_into)
    times_em_tm_into.append(time_em_tm_into)
    iterations_em_tm_into.append(iteration_em_tm_into)
    
    _, segmented_image_em_lp_into, time_em_lp_into, iteration_em_lp_into = \
        segmentation_em(test_image_array, new_mask_array, tissue_models, 'label_prop', 'into', tissue_prob_map_ours)
   
    segmented_images_em_lp_into.append(segmented_image_em_lp_into)
    times_em_lp_into.append(time_em_lp_into)
    iterations_em_lp_into.append(iteration_em_lp_into)
    
    
    #***-----------------------------COMPUTE DICE SCORE--------------------------------***#
    # Compute the dice score for em with kmean
    score = dice_score(test_gt_array, segmented_image_em_kmean)
    em_kmean_dice_scores.append(score)
    
    # Compute the dice score for em with tissue model
    score = dice_score(test_gt_array, segmented_image_em_tm)
    em_tm_dice_scores.append(score)
    
    # Compute the dice score for em with label propagation
    score = dice_score(test_gt_array, segmented_image_em_lp)
    em_lp_dice_scores.append(score)
    
    # Compute the dice score for em with label propagation
    score = dice_score(test_gt_array, segmented_image_em_kmean_into)
    em_kmean_into_dice_scores.append(score)
    
    # Compute the dice score for em with tissue model
    score = dice_score(test_gt_array, segmented_image_em_tm_into)
    em_tm_into_dice_scores.append(score)
    
    # Compute the dice score for em with label propagation
    score = dice_score(test_gt_array, segmented_image_em_lp_into)
    em_lp_into_dice_scores.append(score)
```

```python
# Define the path of the test images, labels, masks and outputs
MAIN_PATH = Path('../Lab_03/').resolve()
test_set_dir_path     = MAIN_PATH / 'data' / 'test-set'
test_images_dir_path  = test_set_dir_path / 'testing-images'
test_labels_dir_path  = test_set_dir_path / 'testing-labels'
test_masks_dir_path   = test_set_dir_path / 'testing-masks'
test_outputs_dir_path = test_set_dir_path / 'testing-outputs'


# Define the path of the tissue probability map
tissue_prob_maps_path_mni = test_outputs_dir_path / 'testing-mni-atlas'


# Define the tissue model for segmentation
tissue_models = gt_histograms_posterior
tissue_models[:, np.argmax(tissue_models, 0) == 0] = np.ones((4, 1))
tissue_models = tissue_models[1:, :]
tissue_models = tissue_models / np.sum(tissue_models, 0)[None, :]


#Define sum blank list to store segmented images
mni_segmented_images_em_lp      = []

mni_segmented_images_em_kmean_into = []
mni_segmented_images_em_tm_into    = []
mni_segmented_images_em_lp_into    = []


mni_em_lp_dice_scores         = []

mni_em_kmean_into_dice_scores = []
mni_em_tm_into_dice_scores    = []
mni_em_lp_into_dice_scores    = []


mni_times_em_lp         = []

mni_times_em_kmean_into = []
mni_times_em_tm_into    = []
mni_times_em_lp_into    = []

mni_iterations_em_lp         = []

mni_iterations_em_kmean_into = []
mni_iterations_em_tm_into    = []
mni_iterations_em_lp_into    = []

# Iterate over the test images for the segmentation using tissue model
for k, test_image_path in enumerate(list(test_images_dir_path.iterdir())):
    test_image_name = test_image_path.name.rstrip('.nii.gz')

    # Load test image, label, mask and tissue probability map
    test_image = sitk.ReadImage(str(test_image_path))
    test_gt    = sitk.ReadImage(str(test_labels_dir_path / f'{test_image_name}_3C.nii.gz'))
    test_mask  = sitk.ReadImage(str(test_masks_dir_path / f'{test_image_name}_1C.nii.gz'))
    tissue_prob_map_mni = sitk.ReadImage(str(tissue_prob_maps_path_ours / f'{test_image_name}_atlas.nii.gz'))
    
    # Convert test image, label, and mask into arrays
    test_image_array = sitk.GetArrayFromImage(test_image)
    test_image_array = ((test_image_array - np.min(test_image_array)) / (np.max(test_image_array) - np.min(test_image_array)) * 255).astype(np.uint8)
    test_gt_array    = sitk.GetArrayFromImage(test_gt).astype(np.uint8)
    test_mask_array  = sitk.GetArrayFromImage(test_mask)
    
    tissue_prob_map_mni = sitk.GetArrayFromImage(tissue_prob_map_mni)
    tissue_prob_map_mni = np.where(tissue_prob_map_mni > 1, 1, tissue_prob_map_mni)
    tissue_prob_map_mni = np.where(tissue_prob_map_mni < 0, 0, tissue_prob_map_mni)

    # Read the upadated mask and convert into array 
    new_mask         = sitk.ReadImage(str(test_set_dir_path / 'new-masks'/ f'{test_image_name}brain_mask.nii'))
    new_mask_array   = sitk.GetArrayFromImage(new_mask) 
    

    
    #**---------------------SEGMENTATION USING EM & LABEL PROPAGATION WITH MNI ATLAS -------------------**#
    mni_prob_em_lp, mni_segmented_image_em_lp, mni_time_em_lp, mni_iteration_em_lp = segmentation_em(test_image_array, new_mask_array, tissue_models, 'label_prop', None, tissue_prob_map_mni)
    
    mni_segmented_images_em_lp.append(mni_segmented_image_em_lp)
    mni_times_em_lp.append(mni_time_em_lp)
    mni_iterations_em_lp.append(mni_iteration_em_lp)

    
    # -------------------------------- SEGMENTATION USING EM WITH MNI ATLAS INTO --------------------------------------**#
    _, mni_segmented_image_em_kmean_into, mni_time_em_kmean_into, mni_iteration_em_kmean_into = \
       segmentation_em(test_image_array, new_mask_array, None, 'kmeans', 'into', tissue_prob_map_mni)
    
    mni_segmented_images_em_kmean_into.append(mni_segmented_image_em_kmean_into)
    mni_times_em_kmean_into.append(mni_time_em_kmean_into)
    mni_iterations_em_kmean_into.append(mni_iteration_em_kmean_into)
 
    _, mni_segmented_image_em_tm_into, mni_time_em_tm_into, mni_iteration_em_tm_into = \
        segmentation_em(test_image_array, new_mask_array, tissue_models, 'tissue_models', 'into', tissue_prob_map_mni)
    
    mni_segmented_images_em_tm_into.append(mni_segmented_image_em_tm_into)
    mni_times_em_tm_into.append(mni_time_em_tm_into)
    mni_iterations_em_tm_into.append(mni_iteration_em_tm_into)
    
    _, mni_segmented_image_em_lp_into, mni_time_em_lp_into, mni_iteration_em_lp_into = \
        segmentation_em(test_image_array, new_mask_array, tissue_models, 'label_prop', 'into', tissue_prob_map_mni)
   
    mni_segmented_images_em_lp_into.append(mni_segmented_image_em_lp_into)
    mni_times_em_lp_into.append(mni_time_em_lp_into)
    mni_iterations_em_lp_into.append(mni_iteration_em_lp_into)
    
    
    #***-----------------------------COMPUTE DICE SCORE--------------------------------***#
    # Compute the dice score for em with label propagation and mni atlas
    score = dice_score(test_gt_array, mni_segmented_image_em_lp)
    mni_em_lp_dice_scores.append(score)
    
    # Compute the dice score for em with label propagation
    score = dice_score(test_gt_array, mni_segmented_image_em_kmean_into)
    mni_em_kmean_into_dice_scores.append(score)
    
    # Compute the dice score for em with tissue model
    score = dice_score(test_gt_array, mni_segmented_image_em_tm_into)
    mni_em_tm_into_dice_scores.append(score)
    
    # Compute the dice score for em with label propagation
    score = dice_score(test_gt_array, mni_segmented_image_em_lp_into)
    mni_em_lp_into_dice_scores.append(score)
    

```
